```python
import pandas as pd
from ntscraper import Nitter
```

```python
scraper=Nitter()
```

Nitter has

```python
tweets = scraper.get_tweets("coding",mode='hashtag', number=1000)
```

```python
tweets
```

```python
tweets=tweets['tweets']
```

```python
filtered_data = []

for tweet in tweets:
  filtered_tweet = {}

  filtered_tweet['content'] = tweet['text']
  filtered_tweet['username'] = tweet['user']['username']
  filtered_tweet['likes'] = tweet['stats']['likes']

  filtered_data.append(filtered_tweet)
```

```python
filtered_data
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(filtered_data)
```

```python
df.to_csv("filtered_data.csv", index=False)
```
